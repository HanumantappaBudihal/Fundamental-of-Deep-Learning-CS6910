```python
import wandb
from matplotlib import pyplot as plt
from tensorflow.keras.datasets.fashion_mnist import load_data
```

```python
(x_train, y_train), (x_test, y_test) = load_data()
names = ["T-shirt", "Trouser", "Pullover", "Dress", "Coat", "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
item, imgs = 0, []
f = plt.figure(figsize=(10,5))

for img, label in zip(x_train, y_train):
    if int(label) == item:
        f.add_subplot(2, 5, label+1)
        plt.grid(False)
        plt.imshow(img, cmap="gray")
        plt.xlabel(names[label])
        imgs.append(img)
        item += 1
        if item >= 10:
            break

plt.show(block=True)
```

```python
run = wandb.init(project="IITM-CS6910-Projects",entity="cs21m070_cs21m022")
wandb.log({"Q1": [wandb.Image(img, caption=name) for img, name in zip(imgs,names)]})
run.finish()
```
